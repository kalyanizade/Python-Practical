## Data Type

```python
 # In Python, data types define the type of value a variable can store.


# A data type defines:

       # What kind of value a variable holds
       # What operations can be performed on it
       # How it is stored in memory


##  type(-) :- use to check the witch data type is there 


# 1.Numeric Data Types :-

        # int (Integer) -    # Whole numbers
                             # Unlimited size (no fixed limit)

        # float -           # Decimal numbers
                            # Stored in double precision

        # complex -    # Used in scientific calculations

# 2.Boolean (bool) :- 

                   # Only two values: True, False

# 3.Sequence Data Types :-
                    # String (str) :-  # Immutable
                                       # Indexed
                                       # Supports slicing
            
                    # List    :-  # Ordered
                                 # Mutable
                                 # Allows duplicates
                                 # Can store mixed types
            
                    # Tuple   :- # Ordered
                                 # Immutable
                                 # Faster than list
# 4. Set :- 
                    # Unordered
                    # No duplicates
                    # Mutable
                    # Cannot store list/dict inside

# 5. Dictionary (dict) :-  

                    # Key-value pairs
                    # Keys must be unique
                    # Mutable

######## Mutable vs   Immutable

                 #   | Mutable | Immutable |
                 #   | ------- | --------- |
                 #   | list    | int       |
                 #   | set     | float     |
                 #   | dict    | str       |
                 #   |         | tuple     |
                 #   |         | bool      |


        
                
```

```python
a = 100    # type(a) use to check the witch data type is there 

b = -500
print(type(a))
```

```python
print(3 + 2)   # addition(+)
print(3 - 2)   # subtraction(-)
print(3 * 2)   # multiplication(*)
print(3 / 2)   # division(/)
print(3 ** 2)  # exponential(**)
print(3 % 2)   # modulus(%)
print(3 // 2)  # Floor division operator(//)
```

```python
print(type(10))                  # Int
print(type(3.14))                # Float
print(type(1 + 3j))              # Complex
print(type('KalyaniZade'))       # String
print(type([1, 2, 3]))           # List
print(type({'name':'kalyani'}))  # Dictionary
print(type({9.8, 3.14, 2.7}))    # Set
print(type((9.8, 3.14, 2.7)))    # Tuple
print(type(3 == 3))              # Bool
print(type(3 >= 3))              # Bool
```
