 ##  Main Data Structures in Python
              # Python has 4 built-in data structures:
                    # List
                    # Tuple
                    # Set
                    # Dictionary

```python
    # |  Data Structure  |  Ordered   |  Mutable   |  Duplicates   |
    #   --------------     -------      -------       ----------     
    # | List            |   ✅ Yes   |  ✅ Yes   |   ✅ Yes      |
    # | Tuple           |   ✅ Yes   |  ❌ No    |   ✅ Yes      |
    # | Set             |  ❌ No    |  ✅ Yes    |   ❌ No       |
    # | Dictionary     |  ✅ Yes    |   ✅ Yes   |   Keys ❌     |
```

```python
# Diff betn:= 
            
            #  |   Feature           | List  | Tuple    | Set      |       Dict        |
            # ------------------------------------------------------------------------------ 
            #  |  Ordered           | ✅    |  ✅     |   ❌     |     ✅            |
            #  |  Mutable           | ✅    |  ❌     |   ✅     |    ✅             |
            #  |  Allows Duplicates | ✅    |  ✅     |   ❌     |   Keys ❌        |
            #  |  Indexed           | ✅    |  ✅     |   ❌    |    By key         |
            #  |  Syntax            | `[]`  | `()`    |   `{}`   | ` {key:value}`    |
```
