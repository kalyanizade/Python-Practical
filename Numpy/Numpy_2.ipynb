# 1. Array Creation Functions

```python
import numpy as np
```

```python
a = np.array([1,2,3])      # Create an array from a list(- np.arange(start, stop, step) → you specify the step size)
print("Array a:" , a)
```

```python
# Create an array with evenly spaced values 
b = np.arange(0,10,2)     # Values from 0 to 10 with step 2
print("Array b:" , b)
```

```python
# Create an array with linearly spaced values
c = np.linspace(0,1,5)     # 5 values evenly spaced between 0 and 1 ( np.linspace(start, stop, num) → you specify the number of points)
print("Array c:" , c)
```

```python
# Create an array filled with zeros
d = np.zeros((2, 3))  # 2x3 array of "zeros"
print("Array d:\n" , d)
```

```python
# Create an array filled with ones
e = np.ones((3, 2))  # 3 x 2  array of "ones"
print("Array d:\n" , e)
```

```python
# Create an identity matrix
f = np.eye(4)       # 4x4 identity matrix
print("Indentity matrix f:\n" , f)
```

# 2. Array Manipulation Functions

```python
# Reshape an array
a1 = np.array([1,2,3])
reshape = np.reshape(a1, (1,3))     # Reshape to 1x3
print("Reshape array:" , reshape)
```

```python
# Flatten an array
f1 = np.array([[1,2], [3,4]])
flattened = np.ravel(f1)    # ravel() converts a multi-dimensional array into a 1D array.
print("Flattened array:", flattened)
```

```python
# Transpose an array
e1 = np.array([[1, 2], [3, 4]])
transposed = np.transpose(e1)  # Converting rows into columns and columns into rows.
print("Transposed array:\n", transposed)
```

```python
# Stack arrays vertically
a2 = np.array([1, 2])
b2 = np.array([3, 4])
stacked = np.vstack([a2, b2])  # Stack a and b vertically (stacks arrays row-wise (one below another).)
print("Stacked arrays:\n", stacked)
```

# 4. Statistical Functions

```python
s = np.array([1,2,3,4])
mean = np.mean(s)
print("Meanof s:" , mean)

# mean = sum of all element / number of element = 1+2+3+4/4 = 10/4 = 2.5
```

```python
# Standard deviation of an array
std_dev = np.std(s)
print("Standard deviation of s:", std_dev)
```

```python
# Minimum element of an array
minimum = np.min(s)
print("Min of s:", minimum)
```

```python
# Maximum element of an array
maximum = np.max(s)
print("Max of s:", maximum)
```

# 7 Boolean & Logical Functions

```python
# Check if all elements are True
# all
logical_test = np.array([True, False, True])
all_true = np.all(logical_test)  # Check if all are True
print("All elements True:", all_true)
```

```python
# Check if all elements are True
logical_test = np.array([True, False, True])
all_true = np.all(logical_test)  # Check if all are True
print("All elements True:", all_true)
```

```python
# Check if all elements are True
logical_test = np.array([False, False, False])
all_true = np.all(logical_test)  # Check if all are True
print("All elements True:", all_true)
```

```python
# Check if any elements are True
# any
any_true = np.any(logical_test)  # Check if any are True
print("Any elements True:", any_true)
```

# 8. Set Operations

```python
# Intersection of two arrays
set_a = np.array([1, 2, 3, 4])
set_b = np.array([3, 4, 5, 6])
intersection = np.intersect1d(set_a, set_b)
print("Intersection of a and b:", intersection)
```

```python
# Union of two arrays
union = np.union1d(set_a, set_b)
print("Union of a and b:", union)
```

# 9. Array Attribute Functions

```python
# Array attributes
a = np.array([1, 2, 3])
shape = a.shape  # Shape of the array
size = a.size    # Number of elements
dimensions = a.ndim  # Number of dimensions
dtype = a.dtype   # Data type of the array

print("Shape of a:", shape)
print("Size of a:", size)
print("Number of dimensions of a:", dimensions)
print("Data type of a:", dtype)
```
