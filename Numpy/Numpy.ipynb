```python
# NumPy arrays are usually C-contiguous by default.

# order='c' :=
            # C → Computer → Row-wise  
            # 👉 order='C' means C-style order (Row-major order).
            # 👉 It fills the new array row by row.

        #  ex:= [1 2 3 4 5 6]  
        #          [[1 2 3]
        #          [4 5 6]]

# order='F' :=
            # F → Fortran → Column-wise
            # 👉 order='F' means Fortran-style order (Column-major order)
            # 👉 It fills the array column by column.

        #  ex:=  [1 2 3 4 5 6]
        #             [[1 3 5]
        #             [2 4 6]]

# order='A':= 
            # 'A' = “Adjust according to original memory layout”
            # 👉 order='A' means “Any” order
            # 👉It depends on how the array is stored in memory.
            #  If the array is stored in order:C (row-major) then it behaves like 'C'
            #  If the array is stored in order:f (column-major) then it behaves like 'f'


# Difference Between 'C' and 'F'

        # | Order | Meaning       | Filling Style                         |
        # | ----- | ------------- | -----------------------------------   |
        # | `'C'` | C-style       | Row-wise                              |
        # | `'F'` | Fortran-style | Column-wise                           |
        # | `'A'` |  Automatic    | Uses array’s existing memory order    |
```

```python
arr.reshape(2,3,order='c')   # order='C' means C-style order (Row-major order).
```

```python
arr.reshape(2,3,order='F')   # order='F' means Fortran-style order (Column-major order)
```

```python
arr.reshape(2,3,order='A')  # order='A' almost give you c type output 
```

# python program to generat otp 

```python
import random

def generate_otp(length=4):      # Creates a function with default OTP length = 4.
    """Generate a numeric OTP of a specified length."""
    
    digits = '012345'      # This will generate OTP using only digits:
    otp = ''.join(random.choice(digits) for _ in range(length))   # Joins them into one string
    
    return otp


otp_length = 4  # Creates a function with OTP length = 4.

otp = generate_otp(otp_length)
print(f"Your OTP is: {otp}")

```
