## islower()

```python
    ## islower() checks whether all alphabet characters in a string are lowercase.

        # üëâ Returns True if all letters are lowercase
        # üëâ Returns False otherwise
        # üëâ Numbers and spaces are ignored.
        # üëâ It only checks alphabet characters.





     ## Important Rules

                    # ‚úî At least one alphabet must be present
                    # ‚úî All alphabets must be lowercase
                    # ‚úî Numbers and spaces don‚Äôt affect result
                    # ‚ùå Empty string returns False


     ### diff betn  islower() and  isupper()

               ## | Method      | Checks                    |
                # | ----------- | ------------------------- |
                # | `islower()` | All letters are lowercase |
                # | `isupper()` | All letters are uppercase |



     ### diff betn  lower() and  islower()

            # | Feature          | `lower()`Action (convert) |`islower()`Question (check)   |
        # --------------- -------------- ---------------------  -------------------- 
            # | Type             |  String method            |  String method               |
            # | Purpose          |  Converts to lowercase    |  Checks if lowercase         |
            # | Return Type      |  String                   |  Boolean (True/False)        |
            # | Modifies String  |  Returns new string       |  No modification             |
```

```python
# islower():Checks if all alphabets in a string are lowercase

challenge = 'thirty days of python'
print(challenge.islower()) # True
challenge = 'Thirty days of python'
print(challenge.islower()) # False
```

```python
print("python".islower())           # True
print("thirty days".islower())      # True
print("python123".islower())        # True (numbers ignored)
```

```python
print("123".islower())   # False (no alphabet present)
```

## isupper()

```python
    ## isupper() checks whether all alphabet characters in a string are uppercase.
        
        # üëâ Returns True if all letters are uppercase
        # üëâ Returns False otherwise
        # üëâ Numbers and spaces are ignored.
        # üëâ It checks only alphabet characters.

Important Rules

‚úî At least one alphabet must be present
‚úî All letters must be uppercase
‚úî Numbers and spaces do NOT affect result
‚ùå Empty string returns False
```

```python
# isupper(): returns if all characters are uppercase characters

challenge = 'thirty days of python'
print(challenge.isupper()) #  False
challenge = 'THIRTY DAYS OF PYTHON'
print(challenge.isupper()) # True

```

```python
print("PYTHON".isupper())        # True
print("HELLO WORLD".isupper())   # True
print("PYTHON123".isupper())     # True (numbers ignored)
```

```python
print("Python".isupper())   # False (lowercase letters exist)
print("python".isupper())   # False
print("123".isupper())      # False (no alphabet present)
print("".isupper())         # False (empty string)
```

```python
print("HELLO123!".isupper())  # True Even with numbers and symbols,it returns True because all letters are uppercase.
```

## isnumeric()

```python
## isnumeric() checks whether all characters in a string are numeric characters.

            # üëâ Returns True if all characters are numbers
            # üëâ Returns False otherwise


    ## important Rules
        
        # ‚úî Only numeric characters allowed
        # ‚úî No spaces
        # ‚úî No decimal point .
        # ‚úî No negative sign -
        # ‚úî Empty string ‚Üí False

## isdecimal ‚äÇ isdigit ‚äÇ isnumeric {isnumeric() is the most powerful check.}
```

```python
print("123.45".isnumeric())   # False (decimal point not allowed)
print("123a".isnumeric())     # False
print("".isnumeric())         # False
```

```python
print("¬Ω".isdecimal())   # False
print("¬Ω".isdigit())     # False
print("¬Ω".isnumeric())   # True
```

## swapcase()

```python
## swapcase() is used to change lowercase letters to uppercase and uppercase letters to lowercase.

            # üëâ It reverses the case of each alphabet character.
            # üëâ Returns a new string (does not modify original).

    ## Important Points :- 

           # ‚úî Lowercase ‚Üí Uppercase
           # ‚úî Uppercase ‚Üí Lowercase
           # ‚úî Numbers & symbols remain unchanged
           # ‚úî Returns new string

   ## üîÑ Difference Between Case Methods
        
                    # | Method       | Work                         |
        ## ------------ | ----------------------------------------------- ##
                    # | lower()      | Convert all to lowercase     |
                    # | upper()      | Convert all to uppercase     |
                    # | title()      | Capitalize each word         |
                    #| capitalize() | Capitalize first letter only |
                    #| swapcase()   | Reverse the case             |
```

```python
# swapcase(): Checks if String Starts with the Specified String
  
challenge = 'thirty days of python'
print(challenge.swapcase())   # THIRTY DAYS OF PYTHON
challenge = 'Thirty Days Of Python'
print(challenge.swapcase())  # tHIRTY dAYS oF pYTHON
```

## startswith() 

```python
## startswith() checks whether a string starts with a specified prefix.

# üëâ Returns True if the string begins with the given value
# üëâ Otherwise returns False

## SYNTAX -- string.startswith(prefix, start, end):=
                        # prefix ‚Üí string (or tuple of strings) to check
                        # start ‚Üí optional starting index
                        # end ‚Üí optional ending index

## Important Points :-

                # ‚úî Case-sensitive
                # ‚úî Returns Boolean (True/False)
                #‚úî Does NOT modify string


        # |   Method       |    Checks              |
        # | ------------   | -------------------   |
        # | startswith()   |  Beginning of string  |
        # | endswith()     |  End of string        |
```

```python
# startswith(): Checks if String Starts with the Specified String

challenge = 'thirty days of python'
print(challenge.startswith('thirty')) # True
challenge = '30 days of python'
print(challenge.startswith('thirty')) # False
```

```python
text = "thirty days of python"

print(text.startswith("thirty"))  # True
print(text.startswith("days"))    # False
```

```python
text = "thirty days of python"

print(text.startswith("days", 7))  # True
```

```python
text = "Python is easy"        # üëâ Returns True if it matches any one of them.

print(text.startswith(("Java", "Python")))  # True
```

```python
print("Python".startswith("p"))  # False üëâ Because it is case-sensitive.
```
